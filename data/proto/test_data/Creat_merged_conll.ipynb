```python
import os
import random
import argparse
```

```python
def load_dataset(path_dataset):
    """Load dataset into memory from text file"""
    dataset = []
    with open(path_dataset) as f:
        words, tags = [], []
        # Each line of the file corresponds to one word and tag
        for line in f:
            if line != '\n':
                line = line.strip('\n')
                if len(line.split()) > 1:
                    word = line.split()[0]
                    #tag = line.split()[-1]
                else:
                    word = line.split()[0]
                    # print(line)
                    continue
                try:
                    if len(word) > 0:
                        word = str(word)
                        words.append(word)
                        #tags.append(tag)
                except Exception as e:
                    print('An exception was raised, skipping a word: {}'.format(e))
            else:
                if len(words) > 0:
                    #assert len(words) == len(tags)
                    dataset.append((words))
                    words, tags = [], []
    return dataset
```

```python
def save_dataset(dataset, save_dir,file_name):
    """Write sentences.txt and tags.txt files in save_dir from dataset

    Args:
        dataset: ([(["a", "cat"], ["O", "O"]), ...])
        save_dir: (string)
    """
    # Create directory if it doesn't exist
    print('Saving in {}...'.format(save_dir))
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    # Export the dataset
    with open(os.path.join(save_dir, file_name), 'w') as file_sentences:
        for words in dataset:
            file_sentences.write('{}\n'.format(' '.join(words)))
            #file_tags.write('{}\n'.format(' '.join(tags)))
    print('- done.')
```

```python
save_dir = 'Conll_Format_merged'
import glob
filenames = glob.glob('Conll_Format/*.txt')
for file in filenames:
    dataset = load_dataset(file)
    filename = file.split('/')[1]
    save_dataset(dataset,save_dir,filename)
```
